```python
# Imports

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import graphviz
import networkx as nx
```

```python
COLORS = [
    '#00B0F0',
    '#FF0000'
]
```

```python
# Define the graph
DAG1 = """graph [
directed 1

node [
    id 0
    label "Market Value"
    ]

node [
    id 1
    label "R&D Exp's"
    ]
    
node [
    id 2
    label "Sales"
    ]

node [
    id 3
    label "Tech. Spillovers"
    ]

node [
    id 4
    label "Product Spillovers"
    ]

node [
    id 5
    label "Patent Count"
    ]

node [
    id 7
    label "Capital"
    ]

node [
    id 8
    label "Tax Credits"
    ]

edge [
    source 8
    target 3
    ]
    
edge [
    source 8
    target 4
    ]
 
edge [
    source 3
    target 7
    ]

edge [
    source 3
    target 5
    ]

edge [
    source 3
    target 1
    ]

edge [
    source 4
    target 2
    ]

edge [
    source 5
    target 0
    ]

edge [
    source 1
    target 5
    ]


edge [
    source 1
    target 7
    ]
 

edge [
    source 7
    target 2
    ]
 
 
 

edge [
    source 2
    target 0
    ]
 

]
    
    """
```

```python
# Get the graph
graph = nx.parse_gml(DAG1)

# Plot
nx.draw(
    G=graph, 
    with_labels=True,
    node_size=1000,
    font_size=8,
    node_color=COLORS[0],
    font_color='black'
)
```
